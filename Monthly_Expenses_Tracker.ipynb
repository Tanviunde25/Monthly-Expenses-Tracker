```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
categories = np.array(['Rent', 'Food', 'Travel', 'Fun', 'Bills'])
budget = np.array([1200, 400, 150, 200, 300])
actual = np.array([1200, 520, 110, 280, 290])
diff = budget - actual
print("Differences per category: ",diff)
```

```python
x = np.arange(len(categories))
width = 0.2
plt.bar(x - 0.15, budget, width, label='Budget')
plt.bar(x + 0.15, actual, width, label='Actual')
plt.xticks(x, categories)
plt.ylabel('Amount ($)')
plt.title('Budget vs Actual')
plt.legend()
plt.show()
```
